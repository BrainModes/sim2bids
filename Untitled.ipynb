### Steps to traversing input files 

1. Receive input files and get all content
2. Check if there are overlapping files 
    2.1 if no overlaps, it's a single subject input. Continue with existing structure in this case.
    2.2 if overlaps, first find directories. If exist, use the scenario 1: folder structure. Otherwise, use scenario 2.

```python


def find_dirs_multi(path, files):
    """
    Find whether there are directories in the input files. 
    """

    for file if files:
        if os.path.isdir(os.path.join(path, file)):
            return True
    return False 

```

```python
files = os.listdir('C:\\Users\\dinar\\Desktop\\New_Connectomes_MRTrix3')
files
```

```python
# infer by numbers
numbers = []

for file in files:
    match = re.findall('^[A-Za-z]{2,}_[0-9]{2,}', file)
    if len(match) > 0:
        numbers.append(match[0])

numbers = list(set(numbers))
numbers
```

```python
def find_matches(paths):
    unique_ids = []
    
    for path in paths:
        match = re.findall('^[A-Za-z]{2,}_[0-9]{2,}', path)
        if len(match) > 0:
            unique_ids.append(match[0])

    return list(set(numbers))
```

```python
subs = OrderedDict()

for idx in range(len(numbers)):   
    subs[numbers[idx]] = [x for x in files if numbers[idx] in x]

subs
```

```python
def get_unique_subs(match):
    subs = OrderedDict()
    
    for idx in range(len(match)):
        subs[match[idx]] = [x for x in match if match[idx] in x]
    
    return subs
```
